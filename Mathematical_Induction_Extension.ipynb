```python
import numpy as np
import matplotlib.pyplot as plt

# Transition matrix
M = np.array([[0.2, 0.8],
              [0.8, 0.2]])

# Initial state vector
initial_state = np.array([[0.7], [0.3]])

# Number of iterations
num_iterations = 25

# List to store the results of each iteration
results = []

# Perform iterations
current_state = initial_state
for i in range(num_iterations):
    current_state = np.dot(M, current_state)
    results.append(current_state.flatten())

# Plot
results = np.array(results)
plt.plot(results[:, 0], label='Probability of being selected')
plt.plot(results[:, 1], label='Probability of not being selected')
plt.xlabel('Iteration')
plt.ylabel('Probability')
plt.title('Convergence of probabilities over iterations')
plt.legend()
plt.show()
```
